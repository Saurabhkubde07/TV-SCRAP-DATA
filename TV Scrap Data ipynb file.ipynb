```python
page_count = 44
while page_count < 44:
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    jobs = soup.find_all('div', class_="tUxRFH")

    for job in jobs:
        TV_Name_=job.find('div',class_="KzDlHZ").text
        Ratings_=job.find('div',class_="XQDdHH").text 
        price_=job.find('div',class_="Nx9bqj _4b5DiR").text
        original_price_ = job.find('div',class_="yRaY8j ZYYwLA").text
        discount_element_ = job.find('div',class_="UkUFwK").text
        features =  job.find_all('li')
        features
        features_1 = []
        for i in job.find_all('li'):
                features_1.append(i.text)
                
                
    
        dict_['TV_Name'].append(TV_Name_)
        dict_['Ratings'].append(Ratings_)
        dict_['price'].append(price_)
        dict_['original_price'].append(original_price_)
        dict_['discount_element'].append(discount_element_)
        dict_['features'].append(features_1)


    # Click on Next Button for change the page
    try:
        next_button = driver.find_element(By.XPATH, "/html/body/div/div/div[3]/div[1]/div[2]/div[26]/div/div/nav/a[11]/span")
        next_button.click()
        time.sleep(3)  # Wait for the page to load
        page_count = page_count + 1

    except NoSuchElementException:
        # If there's no "Next" button, break out of the loop
        break
```
